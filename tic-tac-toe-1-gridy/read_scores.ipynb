```python
import re
```

```python
file_names = [
	# 'score_board_1-1000.log',
	# 'score_board_1001-1999.log',
	# 'score_board_2000-2998.log',
	# 'score_board_2999-3997.log',
	# 'score_board_3998-4996.log',
	# 'score_board_4997-5995.log',
	# 'score_board_5996-6994.log',
	# 'score_board_6995-7993.log',
	'score_board_FINAL.log',
]
_pattern = r'{(.*?) (.*?) (random_.|random|[+-].*? )?([+-].*? )?([+-].*? )?.*?} w: (.*?) d: (.*?) l: (.*)'
_grater_then_random = 0
_win_rates = []
_loose_rates = []

_first_strategy_counters = dict()
_second_strategy_counters = dict()
_third_strategy_counters = dict()

_one_strategy_counters = (0, 0, 0, 0)
_two_strategy_counters = (0, 0, 0, 0)
_three_strategy_counters = (0, 0, 0, 0)

for file_name in file_names:
	with open(file_name) as file:
		random_is_gone = False
		for line in file.readlines():
			match = re.match(_pattern, line)
			if match[1] == 'RandomPlayer':
				random_is_gone = True
			if not random_is_gone:
				_grater_then_random += 1
			
			player_id = match[2]
			first_strategy = match[3].strip()
			second_strategy = match[4].strip() if match[4] is not None else ''
			third_strategy = match[5].strip() if match[5] is not None else ''
			win_count = int(match[6])
			draw_count = int(match[7])
			loose_count = int(match[8])
			total_count = sum([win_count, draw_count, loose_count])
			win_rate = win_count / total_count
			loose_rate = loose_count / total_count

			_win_rates.append((player_id, win_rate, first_strategy, second_strategy, third_strategy))
			_loose_rates.append((player_id, loose_rate, first_strategy, second_strategy, third_strategy))
			if first_strategy in _first_strategy_counters:
				_first_strategy_counters[first_strategy][0] += win_count 
				_first_strategy_counters[first_strategy][1] += loose_count 
				_first_strategy_counters[first_strategy][2] += total_count
				_first_strategy_counters[first_strategy][3].append(win_rate)
				_first_strategy_counters[first_strategy][4].append(loose_rate)
			else:
				_first_strategy_counters[first_strategy] = [win_count, loose_count, total_count, [win_rate], [loose_rate]]
			
			if second_strategy in _second_strategy_counters:
				_second_strategy_counters[second_strategy][0] += win_count 
				_second_strategy_counters[second_strategy][1] += loose_count 
				_second_strategy_counters[second_strategy][2] += total_count
				_second_strategy_counters[second_strategy][3].append(win_rate)
				_second_strategy_counters[second_strategy][4].append(loose_rate)
			else:
				_second_strategy_counters[second_strategy] = [win_count, loose_count, total_count, [win_rate], [loose_rate]]
				
			if third_strategy in _third_strategy_counters:
				_third_strategy_counters[third_strategy][0] += win_count 
				_third_strategy_counters[third_strategy][1] += loose_count 
				_third_strategy_counters[third_strategy][2] += total_count
				_third_strategy_counters[third_strategy][3].append(win_rate)
				_third_strategy_counters[third_strategy][4].append(loose_rate)
			else:
				_third_strategy_counters[third_strategy] = [win_count, loose_count, total_count, [win_rate], [loose_rate]]

print(_grater_then_random)
```

```python
print(*sorted(_loose_rates, key=lambda x: x[1], reverse=True), sep='\n')
```

```python
print(*sorted(_win_rates, key=lambda x: x[1], reverse=True), sep='\n')
```
